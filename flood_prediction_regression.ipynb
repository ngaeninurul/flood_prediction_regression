```python
# Cek outliers
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

for feature in df_train.columns:
    plt.figure(figsize=(10, 6))
    sns.boxplot(x=df_train[feature])
    plt.title(f'Box Plot of {feature}')
    plt.show()
```

```python
# Mengidentifikasi outliers menggunakan IQR
Q1 = df_train.quantile(0.25)
Q3 = df_train.quantile(0.75)
IQR = Q3 - Q1

# Hapus outliers (Filter dataframe untuk hanya menyimpan baris yang tidak mengandung outliers pada kolom numerik)
condition = ~((df_train < (Q1 - 1.5 * IQR)) | (df_train > (Q3 + 1.5 * IQR))).any(axis=1)
df = df_train.loc[condition, df_train.columns]
```

```python
from sklearn.preprocessing import StandardScaler

# Menyimpan data sebelum standardisasi
df_before = df.copy()

# Memastikan hanya data dengan tipe numerikal yang akan diproses
numeric_features = df.select_dtypes(include=['number']).columns
numeric_features

# Standardisasi fitur numerik
scaler = StandardScaler()
df[numeric_features] = scaler.fit_transform(df[numeric_features])
```

```python
# Membuat Histogram Sebelum dan Sesudah Standardisasi
for feature in numeric_features:
    plt.figure(figsize=(12, 6))

    # Histogram Sebelum Standarisasi
    plt.subplot(1, 2, 1)
    sns.histplot(df_before[feature], bins=30, kde=True, color='blue')
    plt.title(f'{feature} Sebelum Standardisasi')
    plt.xlabel(feature)
    plt.ylabel('Frekuensi')

    # Histogram Sesudah Standarisasi
    plt.subplot(1, 2, 2)
    sns.histplot(df[feature], bins=30, kde=True, color='orange')
    plt.title(f'{feature} Sesudah Standardisasi')
    plt.xlabel(feature)
    plt.ylabel('Frekuensi')

    plt.tight_layout()
    plt.show()
```

```python
# Cek duplikasi data
duplicates = df.duplicated()

print("Baris duplikat:")
print(df[duplicates])
```

```python
# Melihat karakteritik data sesudah diolah
df.describe(include='all')
```

Data sudah memiliki distribusi yang cukup baik, hal ini dibuktikan dengan standar deviasi yang mendekati 1 dan rentang nilai yang sama

```python
# Membuat visualisasi distribusi data sesudah diolah

# Menghitung jumlah variabel
num_vars = df.shape[1]

# Menentukan jumlah baris dan kolom untuk grid subplot
n_cols = 4  # Jumlah kolom yang diinginkan
n_rows = -(-num_vars // n_cols)  # Ceiling division untuk menentukan jumlah baris

# Membuat subplot
fig, axes = plt.subplots(n_rows, n_cols, figsize=(20, n_rows * 4))

# Flatten axes array untuk memudahkan iterasi jika diperlukan
axes = axes.flatten()

# Plot setiap variabel
for i, column in enumerate(df.columns):
    df[column].hist(ax=axes[i], bins=20, edgecolor='black')
    axes[i].set_title(column)
    axes[i].set_xlabel('Value')
    axes[i].set_ylabel('Frequency')

# Menghapus subplot yang tidak terpakai (jika ada)
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

# Menyesuaikan layout agar lebih rapi
plt.tight_layout()
plt.show()
```

```python
# Memilih fitur yang memiliki hubungan dengan fitur target (FloodProbability)

# Menghitung korelasi antara variabel target dan semua variabel lainnya
target_corr = df.corr()['FloodProbability']

# (Opsional) Mengurutkan hasil korelasi berdasarkan kekuatan korelasi
target_corr_sorted = target_corr.abs().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
target_corr_sorted.plot(kind='bar')
plt.title(f'Correlation with FloodProbability')
plt.xlabel('Variables')
plt.ylabel('Correlation Coefficient')
plt.show()
```

DATA SPLITTING

```python
import sklearn
from sklearn import datasets

# Memisahkan fitur (X) dan target (y)
X = df.drop(columns=['FloodProbability'])
y = df['FloodProbability']

from sklearn.model_selection import train_test_split

# Membagi dataset menjadi training dan testing
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

# Menghitung panjang/jumlah data
print("Jumlah data: ",len(X))
# Menghitung panjang/jumlah data pada x_test
print("Jumlah data latih: ",len(x_train))
# Menghitung panjang/jumlah data pada x_test
print("Jumlah data test: ",len(x_test))
```

MODELLING

1. LEAST ANGLE REGRESSION (LARS)

```python
# Melatih model
from sklearn import linear_model

lars = linear_model.Lars(n_nonzero_coefs=1).fit(x_train, y_train)
pred_lars = lars.predict(x_test)

# Uji kinerja model dengan metrik regresi
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
mae_lars = mean_absolute_error(y_test, pred_lars)
mse_lars = mean_squared_error(y_test, pred_lars)
r2_lars = r2_score(y_test, pred_lars)

print(f"MAE: {mae_lars}")
print(f"MSE: {mse_lars}")
print(f"RÂ²: {r2_lars}")
```

```python
# Membuat dictionary untuk menyimpan hasil evaluasi
data = {
    'MAE': [mae_lars],
    'MSE': [mse_lars],
    'R2': [r2_lars]
}

# Konversi dictionary menjadi dataframe pengujian
df_results = pd.DataFrame(data, index=['Lars'])
df_results
```

2. LINEAR REGRESSION


```python
# Melatih model
from sklearn.linear_model import LinearRegression

LR = LinearRegression().fit(x_train, y_train)
pred_LR = LR.predict(x_test)

# Uji kinerja model
mae_LR = mean_absolute_error(y_test, pred_LR)
mse_LR = mean_squared_error(y_test, pred_LR)
r2_LR = r2_score(y_test, pred_LR)

print(f"MAE: {mae_LR}")
print(f"MSE: {mse_LR}")
print(f"RÂ²: {r2_LR}")
```

```python
# Tambahkan hasil ke dataframe pengujian
df_results.loc['Linear Regression'] = [mae_LR, mse_LR, r2_LR]
df_results
```

3. GRADIENT BOOSTING REGRESSOR (GBR)

```python
# Melatih model
from sklearn.ensemble import GradientBoostingRegressor

GBR = GradientBoostingRegressor(random_state=184)
GBR.fit(x_train, y_train)
pred_GBR = GBR.predict(x_test)

# Uji kinerja model
mae_GBR = mean_absolute_error(y_test, pred_GBR)
mse_GBR = mean_squared_error(y_test, pred_GBR)
r2_GBR = r2_score(y_test, pred_GBR)

print(f"MAE: {mae_GBR}")
print(f"MSE: {mse_GBR}")
print(f"RÂ²: {r2_GBR}")
```

```python
# Tambahkan hasil ke dataframe pengujian
df_results.loc['GradientBoostingRegressor'] = [mae_GBR, mse_GBR, r2_GBR]
df_results
```

Ketentuan:
1. MAE : harus rendah
2. MSE : harus rendah
3. R^2 : harus tinggi

Kesimpulan:

Model dengan algoritma Linear Regression adalah piihan terbaik
